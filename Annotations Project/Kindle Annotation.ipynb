```python
import json
from datetime import datetime

# Absolute file path
file_path = "/Users/jeremyfeagan/Library/Mobile Documents/com~apple~CloudDocs/MyGitRepo/Chapter 14 Annotations.txt"
# File paths
input_file_path = "/Users/jeremyfeagan/Library/Mobile Documents/com~apple~CloudDocs/MyGitRepo/Chapter 14 Annotations.txt"
output_file_path = '/Users/jeremyfeagan/Library/Mobile Documents/com~apple~CloudDocs/MyGitRepo/Page_Number_Estimation_Test_Chapter14.txt'

def test_page_number_estimation(file_path, output_file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        json_data = json.load(file)
        annotations = json_data['annotations']

    with open(output_file_path, 'w', encoding='utf-8') as output_file:
        for annotation in annotations:
            position = annotation['position']
            page = estimate_page(position)
            output_file.write(f"Position: {position}, Estimated Page: {page}\n\n")

def estimate_page(position):
    # Define known positions and their corresponding page numbers
    known_positions = [131928, 167733, 181139, 483768, 483914, 484201]
    known_pages = [80, 99, 106, 268, 290, 291]

    # Find the two known positions that the given position falls between
    for i in range(len(known_positions) - 1):
        if known_positions[i] <= position <= known_positions[i + 1]:
            # Perform linear interpolation
            pos_start, pos_end = known_positions[i], known_positions[i + 1]
            page_start, page_end = known_pages[i], known_pages[i + 1]

            # Linear interpolation formula
            return page_start + (position - pos_start) * (page_end - page_start) / (pos_end - pos_start)

    # If the position is out of the known range, return an indication
    return "Unknown page"

def extract_context_and_notes_with_diagnostics(file_path):
    today = datetime.now().strftime("%Y-%m-%d")
    output_file_name = f"Annotations_Diagnostics_{today}.txt"

    with open(file_path, 'r', encoding='utf-8') as file:
        json_data = json.load(file)

    annotations = json_data['annotations']

    with open(output_file_name, 'w', encoding='utf-8') as output_file:
        for annotation in annotations:
            position = annotation['position']
            page = estimate_page(position)
            context = annotation.get('context', 'No context')
            note = annotation.get('note')
            note_status = 'Has Note' if note else 'No Note'
            
            diagnostic_text = (
                f"Position: {position}, Estimated Page: {page}\n"
                f"Context: {context}\n"
                f"Note Status: {note_status}\n\n"
            )
            output_file.write(diagnostic_text)

            if note:
                output_text = f"Note: {note}\n\n"
                output_file.write(output_text)

            
# Call the function with the specified file path
extract_context_and_notes_with_diagnostics(file_path)
# Run the test
test_page_number_estimation(input_file_path, output_file_path)
```
