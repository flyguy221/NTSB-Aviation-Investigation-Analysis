```python
#get 2 right
import re
import html

def read_fdf_file(file_path):
    with open(file_path, 'r', encoding='latin-1') as f:
        return f.read()

def process_fdf_objects(content):
    pattern = re.compile(r'(\d+ 0 obj)(.*?)(endobj)', re.DOTALL)
    matches = pattern.findall(content)

    for match in matches:
        obj_id, obj_content, _ = match
        print(f"Object ID: {obj_id}")  # Debugging print
         
        # Extract and clean the contents
        contents_match = re.search(r'<span[^>]*>(.*?)</span>', obj_content, re.DOTALL)
        if contents_match:
            raw_contents = contents_match.group(1)
            # Handle line breaks and decode HTML entities
            cleaned_contents = html.unescape(raw_contents.replace("\\r", "\n").replace("\\n", "\n"))
            cleaned_contents = re.sub(r'\\(.)', r'\1', cleaned_contents)  # Remove backslashes before characters
            print(f"Cleaned Contents: {cleaned_contents}")
        else:
            print("No contents found.")

        # Other fields extraction for completeness
        nm_match = re.search(r'/NM\(([^)]+)\)', obj_content)
        subj_match = re.search(r'/Subj\(([^)]+)\)', obj_content)
        rect_match = re.search(r'/Rect\[(.*?)\]', obj_content)
        irt_match = re.search(r'/IRT\(([^)]+)\)', obj_content)

        if nm_match: print(f"NM: {nm_match.group(1)}")
        if subj_match: print(f"Subj: {subj_match.group(1)}")
        if rect_match: print(f"Rect: {rect_match.group(1)}")
        if irt_match: print(f"IRT: {irt_match.group(1)}")
        print("------")

# Update the file path to your FDF file's actual location
file_path = '/Users/jeremyfeagan/Desktop/adobe fdf.txt'  # Update this to the actual file path
fdf_content = read_fdf_file(file_path)
process_fdf_objects(fdf_content)
```

```python
import re
from html import unescape

def parse_fdf(fdf_filename):
    # Reading the FDF file content
    with open(fdf_filename, 'r', encoding='utf-8') as file:
        fdf_content = file.read()
    
    # Regular expression patterns to find annotations
    highlight_pattern = r'/T\(Quote from text\)/V\((.*?)\)'
    comment_pattern = r'/T\(Paraphrase\)/V\((.*?)\)'
    insight_pattern = r'/T\(Insight\)/V\((.*?)\)'
    
    # Finding and cleaning annotations
    highlights = [unescape(re.sub(r'\\[rn]', '', match)) for match in re.findall(highlight_pattern, fdf_content)]
    comments = [unescape(re.sub(r'\\[rn]', '', match)) for match in re.findall(comment_pattern, fdf_content)]
    insights = [unescape(re.sub(r'\\[rn]', '', match)) for match in re.findall(insight_pattern, fdf_content)]
    
    return highlights, comments, insights

def format_rtf(highlights, comments, insights):
    rtf_content = r'{\rtf1\ansi'
    
    for highlight, comment, insight in zip(highlights, comments, insights):
        # Formatting the highlight
        rtf_content += r'\b\highlight2 QUOTE FROM TEXT: ' + highlight + r'\b0\par'
        # Formatting the comment
        rtf_content += r'\cf1\li200 ' + comment + r'\cf0\li0\par'
        # Formatting the insight
        rtf_content += r'Insight: ' + insight + r'\par'
    
    rtf_content += r'}'
    
    return rtf_content

def save_rtf(rtf_content, output_filename):
    with open(output_filename, 'w', encoding='utf-8') as file:
        file.write(rtf_content)

def main():
    fdf_filename = '/Users/jeremyfeagan/Desktop/adobe fdf.txt'  # Update this to your FDF file path
    output_filename = '/Users/jeremyfeagan/Desktop/Output RTF.rtf'  # Update this to your desired output file name
    
    highlights, comments, insights = parse_fdf(fdf_filename)
    rtf_content = format_rtf(highlights, comments, insights)
    save_rtf(rtf_content, output_filename)
    
    print(f"RTF document '{output_filename}' has been created successfully.")

if __name__ == "__main__":
    main()
```

```python
import re
from html import unescape

# Define a function to parse the FDF file
def parse_fdf(fdf_file_path):
    with open(fdf_file_path, 'r', encoding='utf-8') as file:
        fdf_content = file.read()
    
    # Regular expressions to find annotations
    highlight_pattern = re.compile(r'/Subtype/Highlight.*?/Contents\((.*?)\)', re.DOTALL)
    comment_pattern = re.compile(r'/Subtype/Text.*?/Contents\((.*?)\)', re.DOTALL)
    insight_pattern = re.compile(r'Insight: (.*?)\n', re.DOTALL)
    
    highlights = highlight_pattern.findall(fdf_content)
    comments = comment_pattern.findall(fdf_content)
    insights = insight_pattern.findall(fdf_content)
    
    # Clean and format the extracted content
    formatted_content = ""
    
    for highlight in highlights:
        formatted_content += "\\par \\b \\highlight3 QUOTE FROM TEXT: " + unescape(highlight).replace('\r', '').replace('\n', ' ') + "\\b0\\par\n"
        
        related_comments = [comment for comment in comments if unescape(comment).startswith(highlight)]
        for comment in related_comments:
            formatted_content += "\\par \\cf2 " + "    " + unescape(comment).replace('\r', '').replace('\n', ' ').replace(highlight, '', 1).strip() + "\\cf0\\par\n"
    
    for insight in insights:
        formatted_content += "\\par Insight: " + unescape(insight).replace('\r', '').replace('\n', ' ') + "\\par\n"
    
    return formatted_content

# Define a function to save the formatted content to an RTF file
def save_to_rtf(formatted_content, output_file_path):
    rtf_header = "{\\rtf1\\ansi\\ansicpg1252\\deff0\\nouicompat\\deflang1033{\\fonttbl{\\f0\\fnil\\fcharset0 Calibri;}}\n"
    rtf_header += "{\\colortbl ;\\red0\\green0\\blue255;\\red255\\green255\\blue0;}\n"
    rtf_header += "\\viewkind4\\uc1\\pard\\sa200\\sl276\\slmult1\\f0\\fs22\\lang9 "
    rtf_footer = "}"
    
    with open(output_file_path, 'w', encoding='utf-8') as file:
        file.write(rtf_header + formatted_content + rtf_footer)

# Example usage
fdf_file_path = '/Users/jeremyfeagan/Desktop/adobe fdf.txt'  # Update this to your FDF file path
output_file_path = '/Users/jeremyfeagan/Desktop/Output RTF.rtf'  # Update this to your desired output RTF file path

formatted_content = parse_fdf(fdf_file_path)
save_to_rtf(formatted_content, output_file_path)
print("RTF file has been created successfully.")
```

```python
import re
from html import unescape

# Define a function to strip HTML tags for clean text
def strip_html_tags(text):
    clean = re.compile('<.*?>')
    return unescape(re.sub(clean, '', text))

# Parse the FDF file and organize annotations
def parse_fdf(fdf_filename):
    annotations = []
    with open(fdf_filename, 'r', encoding='utf-8') as f:
        content = f.read()

    # Split the content by annotations, this regex might need adjustments depending on the FDF structure
    items = re.split(r'annots\[\d+\]\.', content)

    for item in items:
        if "T(Quote from text)" in item:
            highlight = re.search(r'V\((.*?)\)', item, re.DOTALL)
            comment = re.search(r'T\(Pasha & Stokes \(2018\)\)(.*?)V\((.*?)\)', item, re.DOTALL)
            insight = re.search(r'T\(Insight\)(.*?)V\((.*?)\)', item, re.DOTALL)

            if highlight:
                highlight_text = strip_html_tags(highlight.group(1).strip())
                annotations.append(f'QUOTE FROM TEXT: {highlight_text}\n')

            if comment:
                comment_text = strip_html_tags(comment.group(2).strip())
                annotations.append(f'\tComment: {comment_text}\n')

            if insight:
                insight_text = strip_html_tags(insight.group(2).strip())
                annotations.append(f'\tInsight: {insight_text}\n')

    return annotations

# Function to save annotations to a file (could be modified for RTF format specifics)
def save_annotations(annotations, output_filename):
    with open(output_filename, 'w', encoding='utf-8') as f:
        for annotation in annotations:
            f.write(annotation + '\n')

# Example usage
fdf_filename = '/Users/jeremyfeagan/Desktop/adobe fdf.txt'  # Update this to your FDF file path
output_filename = '/Users/jeremyfeagan/Desktop/Output RTF03080712.rtf'  # Update this to your desired output file name

annotations = parse_fdf(fdf_filename)
save_annotations(annotations, output_filename)

print(f'Annotations extracted and saved to {output_filename}')
```
